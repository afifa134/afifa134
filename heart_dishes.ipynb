```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/heart_dishes')
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())

# Handle missing values (example: fill with mean for numerical columns)
for col in data.columns:
    if pd.api.types.is_numeric_dtype(data[col]):
        data[col] = data[col].fillna(data[col].mean())
print(data.isnull().sum())
```

```python
data.describe()
```

```python
data.info()
```

```python
# Convert categorical features to numerical using one-hot encoding
categorical_cols = data.select_dtypes(include=['object']).columns
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
```

```python
print(data.head())
```

```python
# Convert boolean columns to numerical (0 and 1)
for col in data.columns:
    if pd.api.types.is_bool_dtype(data[col]):
        data[col] = data[col].astype(int)

print(data.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn_lvq import GlvqModel
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Encode categorical features
categorical_cols = data.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
# Separate features and target
X = data.drop(columns=['target'])
y = data['target']
```

```python
# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
!pip install sklearn-lvq
```

```python
from sklearn_lvq import GlvqModel
from sklearn.metrics import accuracy_score, classification_report

# Initialize the LVQ model
lvq_model = GlvqModel()

# Train the model
lvq_model.fit(X_train, y_train)

# Predict on test data
y_pred = lvq_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
if np.any(np.isnan(X_train)) or np.any(np.isinf(X_train)):
    raise ValueError("X_train contains NaN or infinite values.")
if np.any(np.isnan(y_train)) or np.any(np.isinf(y_train)):
    raise ValueError("y_train contains NaN or infinite values.")
```

```python
import numpy as np

X_train = np.asarray(X_train)
y_train = np.asarray(y_train)

# Flatten y_train if necessary
if y_train.ndim != 1:
    y_train = y_train.ravel()

print("Data converted to NumPy arrays successfully.")
```

```python
if np.any(np.isnan(X_train)) or np.any(np.isinf(X_train)):
    raise ValueError("X_train contains NaN or infinite values.")
if np.any(np.isnan(y_train)) or np.any(np.isinf(y_train)):
    raise ValueError("y_train contains NaN or infinite values.")
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
```

```python
# prompt: random frorest

from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)  # You can adjust hyperparameters here

# Train the model
rf_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred_rf = rf_classifier.predict(X_test)

# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Test Accuracy:", accuracy_rf)
print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
```
